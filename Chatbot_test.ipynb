```python
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core import Settings
from transformers import AutoTokenizer

# Set up the model using Hugging Face's API
model_name = "swap-uniba/LLaMAntino-3-ANITA-8B-Inst-DPO-ITA"
tokenizer = AutoTokenizer.from_pretrained(model_name)

llm = HuggingFaceLLM(
    model_name=model_name,
    tokenizer_name=tokenizer, #Not sure
    context_window=4096,
    max_new_tokens=512,
    generate_kwargs={"temperature": 0.7, "top_k": 50, "top_p": 0.95},
    device_map="auto",  # Uses GPU if available
)

# Set the LLM in LlamaIndex's settings
Settings.llm = llm
```

```python
# References 
# https://github.com/marcopoli/LLaMAntino-3-ANITA/blob/main/README.md
# https://docs.llamaindex.ai/en/stable/module_guides/models/llms/usage_custom/

sys = "Sei un assistente delle vendite retail di prodotti farmaceutici. Rispondi in maniera educata"
messages = [
    {"role": "system", "content": sys},
    {"role": "user", "content": "Qual Ã¨ la capitale d'Italia?"} #Example
]

prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

response = llm.complete(prompt)


print(response)
```
